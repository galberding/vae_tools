```python
from matplotlib import pyplot as pl
import numpy as np
file = np.load('2018-09-14T11:20:24_ngiam.npz')
random_states_rewards = file["random_states_rewards"]
file = np.load('2018-09-14T12:13:01_ngiam.npz')
random_states_rewards = np.append(random_states_rewards,file["random_states_rewards"], axis=0)
file = np.load('2018-09-14T13:12:50_ngiam.npz')
random_states_rewards = np.append(random_states_rewards,file["random_states_rewards"], axis=0)

# for i in range(3):
avg_reward = random_states_rewards[i]
print(avg_reward.shape)
avg_reward_mean = np.sum(np.mean(random_states_rewards, axis=0), axis=-1) / avg_reward.shape[1]
avg_reward_err  = np.std(np.std(random_states_rewards, axis=0), axis=-1)
N = 50
avg_reward_mean = np.convolve(avg_reward_mean, np.ones((N,))/N, mode='valid')
avg_reward_err = np.convolve(avg_reward_err, np.ones((N,))/N, mode='valid')
pl.plot(avg_reward_mean, 'k', color='#CC4F1B', label='JMVAE-Zero')
pl.fill_between(np.arange(len(avg_reward_mean)), avg_reward_mean-avg_reward_err, avg_reward_mean+avg_reward_err,
    alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848')


file = np.load('2018-09-14T14:07:43_tripleloss.npz')
random_states_rewards = file["random_states_rewards"]
file = np.load('2018-09-14T16:34:24_tripleloss.npz')
random_states_rewards = np.append(random_states_rewards,file["random_states_rewards"], axis=0)
file = np.load('2018-09-14T16:35:39_tripleloss.npz')
random_states_rewards = np.append(random_states_rewards,file["random_states_rewards"], axis=0)

# for i in range(3):
avg_reward = random_states_rewards[i]
print(avg_reward.shape)
avg_reward_mean = np.sum(np.mean(random_states_rewards, axis=0), axis=-1) / avg_reward.shape[1]
avg_reward_err  = np.std(np.std(random_states_rewards, axis=0), axis=-1)
N = 50
avg_reward_mean = np.convolve(avg_reward_mean, np.ones((N,))/N, mode='valid')
avg_reward_err = np.convolve(avg_reward_err, np.ones((N,))/N, mode='valid')
pl.plot(avg_reward_mean, 'k', color='#1B2ACC', label='tVAE')
pl.fill_between(np.arange(len(avg_reward_mean)), avg_reward_mean-avg_reward_err, avg_reward_mean+avg_reward_err,
    alpha=0.5, edgecolor='#1B2ACC', facecolor='#089FFF')

file = np.load('2018-09-14T16:15:06_proposed.npz')
random_states_rewards = file["random_states_rewards"]
file = np.load('2018-09-14T16:15:58_proposed.npz')
random_states_rewards = np.append(random_states_rewards,file["random_states_rewards"], axis=0)
file = np.load('2018-09-14T16:16:16_proposed.npz')
random_states_rewards = np.append(random_states_rewards,file["random_states_rewards"], axis=0)

# for i in range(3):
avg_reward = random_states_rewards[i]
print(avg_reward.shape)
avg_reward_mean = np.sum(np.mean(random_states_rewards, axis=0), axis=-1) / avg_reward.shape[1]
avg_reward_err  = np.std(np.std(random_states_rewards, axis=0), axis=-1)
N = 50
avg_reward_mean = np.convolve(avg_reward_mean, np.ones((N,))/N, mode='valid')
avg_reward_err = np.convolve(avg_reward_err, np.ones((N,))/N, mode='valid')
pl.plot(avg_reward_mean, 'k', color='#3F7F4C', label='M$^2$VAE')
pl.fill_between(np.arange(len(avg_reward_mean)), avg_reward_mean-avg_reward_err, avg_reward_mean+avg_reward_err,
    alpha=0.5, edgecolor='#3F7F4C', facecolor='#7EFF99')



pl.ylabel('avg. reward')
pl.xlabel('episode')
pl.legend(loc='lower right', frameon=False)
pl.savefig("reward_during_training.svg")
pl.show()


# Staistic
# JMVAE-Zero
successrate = [63.75, 68.1380952381, 66.1047619048]
print("JMVAE-Zerosuccessrate mean: ", np.mean(successrate))
print("JMVAE-Zerosuccessrate std: ", np.std(successrate))

# TripleElbo
successrate = [68.1380952381, 71.4285714286, 73.9428571429]
print("TripleElbo-Zerosuccessrate mean: ", np.mean(successrate))
print("TripleElbo-Zerosuccessrate std: ", np.std(successrate))


# Proposed
successrate = [81.2571428571, 70.8666666667, 72.5761904762]
print("Proposed-Zerosuccessrate mean: ", np.mean(successrate))
print("Proposed-Zerosuccessrate std: ", np.std(successrate))
```
